```python
import os

source_folder = 'source'
target_folder = 'target'

source_files = os.listdir(source_folder)
target_files = os.listdir(target_folder)

# Count the number of matching filenames
matching_count = sum(1 for file_name in source_files if file_name in target_files)

print(f"Number of matching filenames: {matching_count}")
```

```python
import json

data = []
with open('prompt.json', 'rt') as f:
    for line in f:
        data.append(json.loads(line))

import os

folder = 'source'  # Replace with the actual path to the folder

existing_count = 0

for filename in data:
    filename = filename['source']
    file_path = os.path.join(folder, filename)
    if os.path.exists(file_path):
        existing_count += 1

print(f"Number of existing filenames: {existing_count}")
```
